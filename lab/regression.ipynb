```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import os
```

```python
laptop_price_path = os.path.join("..", "resources", "regression", "linear", "laptop_price.csv")
laptop_price = pd.read_csv(laptop_price_path)
```

```python
laptop_price
```

```python
laptop_price.columns
```

```python
X = laptop_price[['CPU_Frequency (GHz)', 'RAM (GB)']]
y = laptop_price['Price (Euro)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_pred, color='blue')
plt.title("Linear Regression")
plt.xlabel("Price (Euro)")
plt.ylabel("Predicted Price (Euro)")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')
plt.show()
```
